```python
# 常用的导入方式：
import matplotlib.pyplot as plt
```

```python
x = [3, 5, 6, 8]
y = [5, 8, 6, 9]
```

# 一、绘制线性图：plot

```python
# 1. 绘制线性图：x，y都有
plt.plot(x,y)
plt.show()  # 所有plt画的图，都必须用show才能显示出来！
```

```python
# 2. 绘制线性图：只有y，则x给默认的从0开始的自然刻度
plt.plot(y)
plt.show()
```

```python
# 3. 绘制线性图：多个y画在一张图上！
x = [3, 5, 6, 8]
y1 = [5, 8, 6, 9]
y2 = [2, 7, 3, 4]

# 画在一起：必须是分开的，不能是plt.plot(x, y1, y2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

# 二、绘制柱状图：bar、barh

```python
x = [3, 5, 6, 8]
y = [5, 8, 6, 9]
```

```python
# 垂向柱状图：
plt.bar(x,y)
plt.show()
```

```python
# 横向柱状图horizontal：x、y轴互换方向！
plt.barh(x,y)
plt.show()
```

```python
# 堆叠柱状图：多个y画在一起
x = [3, 5, 6, 8]
y1 = [5, 8, 6, 9]
y2 = [2, 7, 3, 4]
y3 = [3, 5, 3, 4]

# 和plot一样，但凡是多图画一起的，必定是“多次画，每次画一组数据”！
plt.bar(x, y1)
plt.bar(x, y2, bottom = y1)
plt.bar(x, y3, bottom = y2)  # 注意此时的基底是y2！
plt.show()
```

# 三、绘制散点图：scatter 

```python
# 直接上多组数据：
x = [3, 5, 6, 8]
y1 = [5, 8, 6, 9]
y2 = [2, 7, 3, 4]
y3 = [3, 5, 3, 4]
```

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.scatter(x, y3)  # 保证了x都是一致的！
plt.show()
```

# 四、图标、属性的设置：

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.randn(10)
y = np.cumsum(y, axis = 0)
y
```

```python
# 原折线图：
plt.plot(y)
plt.show()
```

```python
# 1. 进行属性设置：
# 属性可简写：linestyle = ls, linewidth = lw, color = c, marker不能简写
plt.plot( y, linestyle = '--', linewidth = 1, color = 'r', marker = 'o' )  # marker就是把真实数据点标出来！
plt.show()
```

```python
# 2. 读个数据画一起 + 简写的属性设置：
x = np.arange(1,11)
y1 = np.random.randn(10).cumsum()
y2 = np.random.randn(10).cumsum()

plt.plot( x, y1, 'b-o', x, y2, 'r-^')  # 所有的属性简写“必须”写在一个''内！不能但拎出来！
plt.show()
```

```python
# 3. 加标题、纵横坐标：
plt.plot( x, y1, 'b-o', x, y2, 'r-^')  # 所有的属性简写“必须”写在一个''内！不能但拎出来！
plt.title( 'Title', fontsize = 18, y = 1.05 )     # y是指标题离“水平线(y值不变)”的距离倍数！
plt.xlabel( 'x_data', fontsize = 15, y = 1.05 )
plt.ylabel( 'y_data', fontsize = 15, x = 1.05 )   # x同理
plt.show()
```

# 五、图例、刻度的设置：

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = 'T')  # 右边边界点也要！
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# 1. 画到一起，用“图例”区分：用画图例，最好是每组数据都用一次plt.plot
plt.plot(x, y1, 'b-', linewidth = 1, label = 'sin')    # label：图例的名称
plt.plot(x, y2, 'r--', linewidth = 1, label = 'cos')
plt.legend( loc = 0 )  # 选0是自适应最佳位置 
plt.show()
```

```python
# 2. 设置x和y的显示的范围：方式1 —— plt.plot()
plt.plot(x, y1, 'b-', linewidth = 1, label = 'sin')    # label：图例的名称
plt.plot(x, y2, 'r--', linewidth = 1, label = 'cos')
plt.legend( loc = 0 )  # 选0是自适应最佳位置 

plt.xlim( -4, 2 )
plt.ylim( -2, 2 )

plt.show()
```

```python
# 3. 设置x和y的显示的范围：方式2 —— plt.axis( [] ) 里面是一个列表！
plt.plot(x, y1, 'b-', linewidth = 1, label = 'sin')    # label：图例的名称
plt.plot(x, y2, 'r--', linewidth = 1, label = 'cos')
plt.legend( loc = 0 )  # 选0是自适应最佳位置 

plt.axis( [-4, 2, -2, 2] )  # 先x后y

plt.show()
```

```python
# 4. 设置x、y坐标的“刻度”显示：plt.xticks( [要改的原始坐标点],[新的刻度标志] )
plt.plot(x, y1, 'b-', linewidth = 1, label = 'sin')    # label：图例的名称
plt.plot(x, y2, 'r--', linewidth = 1, label = 'cos')
plt.legend( loc = 0 )  # 选0是自适应最佳位置 

plt.axis( [-4, 4, -1.5, 1.5] )  

plt.xticks( [-np.pi, 0, np.pi], [r'$-\pi$', 0, r'$\pi$'] )  # 新的刻度标志支持“LaTeX”符号输入！！

plt.show()
```

# 六、多子图绘制：

```python
x = [1, 2, 3, 4]
y1 = [2, 4, 3, 7]
y2 = [5, 8, 3, 6]
```

```python
plt.subplot(1,2,1)  
plt.plot(x, y1, 'b-o', linewidth = 1); 
plt.subplot(1,2,2)
plt.bar(x, y2, color = 'r', width = 0.7)
plt.show()
```

```python
# 1. 把整个画布拉好看点：plt.figure( figsize = () )
plt.figure( figsize = (8,3) )

plt.subplot(1,2,1)  
plt.plot(x, y1, 'b-o', linewidth = 1); 
plt.subplot(1,2,2)
plt.bar(x, y2, color = 'r', width = 0.7)
plt.show()
```

```python
# 2. 上下/左右“子图数量不等”的绘制：上面3个子图，下面一个大子图！
plt.figure( figsize = (8,7) )

# 对于小子图：所处环境是2x3！
plt.subplot(2,3,1)  
plt.plot(x, y1, 'b-o', linewidth = 1)
plt.subplot(2,3,2)
plt.bar(x, y2, color = 'r', width = 0.7)
plt.subplot(2,3,3)
plt.bar(x, y2, color = 'y', width = 0.7)
# 对于大子图：所处环境是2x1，里面的第2行！ ★★★
plt.subplot(2,1,2)
plt.plot(x, y1, 'g--o', linewidth = 1)

plt.show()
```

```python
# 同理设计一个：左边3个小子图，右边1个大子图 
plt.figure( figsize = (8,7) )

# 对于小子图：其所处环境为3x3
plt.subplot(3,3,1)  
plt.plot(x, y1, 'b-o', linewidth = 1)
plt.subplot(3,3,4)
plt.bar(x, y2, color = 'r', width = 0.7)
plt.subplot(3,3,7)
plt.bar(x, y2, color = 'y', width = 0.7)
# 对于大子图：其所处环境为1x2，里面的第2列！ ★★★
plt.subplot(1,2,2)
plt.plot(x, y1, 'g--o', linewidth = 1)

plt.show()  # 离的太远，因为子图都是3x3里面最左边的那个！当然离右边太远！！
```

# 七、箱型图、直方图、饼图：

1. 箱型图

作用：显示“**一组数据**”的5个关键值的位置（从小到大）：最小值、1/4值、中间值、3/4值、最大值

图像说明：
- 5条横线：从下到上代表那5个值；
- 若有一个圆圈（异常值）：统计是以4的倍数为根据，如果**数据量不是4的倍数**，会出现异常值！

```python
data = [23, 22, 23, 45, 16, 19, 33, 18, 8, 41, 25, 28, 9, 12, 19, 38]  

plt.boxplot(data)
plt.show()  # 
```

2. 直方图

作用：和箱型图一样，都是为了显示数据的分布规律。直方图就是把数组分成几个组，看每个组内有多少数据！

```python
data= np.random.randn(1000)  # 1000个随机数
plt.hist(data, bins = 15)  # 分15大组
plt.show()
```

3. 饼图

作用：显示一组数据，每个数据占“**数值总和**”的百分比

```python
data = [13, 30, 45, 20] 
label_data = ['AA', 'BB', 'CC', 'DD'] 
```

```python
plt.pie( data, labels = label_data, autopct = '%1.1f%%' )  # autopct自带计算并显示百分比，保留1位小数
plt.show()
```

```python
# 让某一组数/某一块饼突出一点：explode = []
data = [13, 30, 45, 20] 
label_data = ['AA', 'BB', 'CC', 'DD'] 
explode_data = [0, 0, 0.2, 0]

plt.pie( data, labels = label_data, explode = explode_data, autopct = '%1.1f%%' )  # autopct自带计算并显示百分比

plt.show()
```

```python
# 一些别的参数：立体阴影、开画的角度
data = [13, 30, 45, 20] 
label_data = ['AA', 'BB', 'CC', 'DD'] 
explode_data = [0, 0, 0.2, 0]

plt.pie( data, labels = label_data, explode = explode_data, autopct = '%1.1f%%', 
         shadow = True, startangle = 180 ) # 立体阴影、开画角度

plt.axis( 'equal' )  #  图片均匀好看！

plt.show()
```

# 八、面向对象绘图：

```python
# 1. 一个总和例子：
x = np.random.randint(1, 11, 30)  # [1, 11]范围内，随机30个整数
y1 = np.random.randint(1, 6, 30)
y2 = y1 + 5

plt.figure( figsize = (8,6) )

plt.scatter( x, y1, label = 'CUG', color = 'g', s = 25, marker = '^' )
plt.scatter( x, y2, label = 'ZJU', color = 'r', s = 25, marker = 'o' )
plt.legend()

# 设置轴的自定义标志：
xlabel_xticks = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten']
plt.xticks( np.arange(1,11), xlabel_xticks, rotation = 70, fontsize = 14 )  

# 基本标签：
plt.grid() # 网格线
plt.xlabel( 'x_data', fontsize = 15 )
plt.ylabel( 'y_data', fontsize = 15 )
plt.title( 'This is the title', fontsize = 16, y = 1.02 )

plt.show()
```

面向对象绘图：一个图
- 先初始化一个对象，然后对这个对象进行操作！

```python
# fig代表figure那个图片/画布，ax代表“绘图对象”！即ax = plt！
fig, ax = plt.subplots( figsize = (7,6) );  # ()括号里没填数字，就表示只画一个图！

ax.scatter( x, y1, label = 'CUG', color = 'g', s = 25, marker = '^' )
ax.scatter( x, y2, label = 'ZJU', color = 'r', s = 25, marker = 'o' )
ax.legend()

# 设置基本标签：多一个set前缀！
ax.set_xlabel( 'x_data', fontsize = 15 )
ax.set_ylabel( 'y_data', fontsize = 15  )
ax.set_title( 'This is the title', fontsize = 16, y = 1.02 )
ax.grid()  

plt.show()
```

面向对象绘图：多个子图 —— 有些不同！

```python
x = [1, 2, 3, 4]
y1 = [4, 5, 7, 8]
y2 = [3, 6, 8, 10]
```

```python
fig, ax = plt.subplots( 2, 2, figsize = (10,8) )  # 前面2,2 —— 两行两列
# 绘子图：ax[][] —— 从0开始！
ax[0][0].bar(x, y1)
ax[0][1].barh(x, y1)
ax[1][0].plot(x, y2)
ax[1][1].scatter(x, y2)

plt.tight_layout()  # 让图别太挤！—— 画大点！

plt.show()
```
