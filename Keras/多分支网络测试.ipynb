```python
import keras
from keras import layers
```

```python
# 用它的权重和框架：
mobile_net = keras.applications.MobileNetV2( input_shape=(224,224,3), include_top = False, weights = 'imagenet' ) 
```

```python
inputs = keras.Input( shape = (224,224,3) )
```

```python
x = mobile_net(inputs)  # 输出的还是一个4维的
x = keras.layers.GlobalAveragePooling2D()(x)
```

进入到分支部分：

```python
# x1 = keras.layers.Dense( 128, activation='relu' )(x)
# out_x1 = keras.layers.Dense( 5, activation='softmax', name = 'out_x1' )(x1) 

out_item = []
for xtmp in range(1000):
    xfz = keras.layers.Dense( 256, activation='relu' )(x)
    xfz = keras.layers.Dropout(0.3)(xfz)
    out_xfz = keras.layers.Dense( 5, activation='softmax', name = 'out_x{}'.format(xtmp+1) )(xfz)
    out_item.append(out_xfz)
```

```python
# 网络合并：
model = keras.Model( inputs = inputs, outputs = out_item )
```
