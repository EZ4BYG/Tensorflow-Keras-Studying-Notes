# 数据读入

```python
import pandas as pd # 用pandas对数据进行预处理是最牛的！
import numpy as np
```

```python
data = pd.read_csv('E:\Python_code\keras_total\日月光华-keras课程资料\小型数据集\Advertising.csv')  # 用绝对路径最好了
```

```python
# 第一列(索引是0)是自动加上的unnamed；TV、radio、newspaper是我们要的特征值x，sales是目标值y
# 读入的多行多列数据，其实当成和numpy一样的“二维数组”即可。
data.head(10)  
```

```python
x = data[ data.columns[1:4] ]  # 用切片取出所需要的3列 —— x
y = data[ data.columns[-1] ]   # 用切片取出最后一列 —— y
# 用 y = data.iloc[:,-1]  和二维数组的切片方式完全一样！

type(y)
```

```python
x.shape()
```

# 模型搭建

```python
import keras
```

```python
# 创建连续模型：
model = keras.Sequential()        
```

```python
# 为模型添加一层一层的堆叠：
from keras import layers
model.add( layers.Dense(1, input_dim = 3) )
# 说明：
# 创建一层Dense层：输出是1个预测数，所以前面是1；输入是3个特征，所以输入维度是3；
# 这样，keras会自定创建3个神经元，即：创建3个权重、1个偏置 = 共4个待训练的参数 —— y_pret = w1*x1 + w2*x2 + w3*x3 + b

model.summary()
```

```python
# 模型编译：还是一个回归问题，因此目标函数损失函数一般仍选均方差(mse)，优化方式选用adam算法。
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
# 模型训练：
model.fit(x, y, epochs=2000)
```

```python
# 预测：
y_pret = model.predict( pd.DataFrame([[300,0,0]]) )
print(y_pret)
```
