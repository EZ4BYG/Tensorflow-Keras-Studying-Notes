```python
import keras
from keras import layers 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# iris.csv数据集：燕尾花的分类
data = pd.read_csv('E:/Python_code/keras_total/日月光华-keras课程资料/小型数据集/iris.csv')
```

```python
data.head()
```

```python
data.Species.unique()
```

```python
# 用一个字典来进行映射的顺序分类：
species_dic = { 'setosa':0, 'versicolor':1, 'virginica':2 }
```

```python
# 利用map函数转换原数据，并重新赋值替代原数据：
data['Species'] = data.Species.map(species_dic)
```

```python
data.head()
```

```python
# 特征x提取：
x = data[ data.columns[1:5] ]
y = data.Species
```

```python
model = keras.Sequential() 
model.add( layers.Dense(3, input_dim = 4, activation = 'softmax') )
```

```python
model.summary()
```

```python
# 对“目标数据”使用顺序编码时，用：sparse_categorical_crossentropy 损失函数，计算softmax交叉熵！
model.compile( optimizer = 'adam',
               loss = 'sparse_categorical_crossentropy',
               metrics = ['acc']
)
```

```python
model.fit(x, y, epochs = 500)
```
